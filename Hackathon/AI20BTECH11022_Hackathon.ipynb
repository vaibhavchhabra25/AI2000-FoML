```python
import numpy as np
import pandas as pd
```

```python
train_df = pd.read_csv('train.csv')
test_df = pd.read_csv('test.csv')
```

```python
train_df.head()
```

```python
test_df.head()
```

```python
test_df = test_df.drop(['Id'], axis=1)  # dropping the 'Id' coumn from test dataset
```

```python
# splitting the 'Crash Date/Time' column to 'Month', 'Date', 'Day of Week', 'Hour'

data = [train_df,test_df]
for i in data:
    i['Crash Date/Time'] = pd.to_datetime(i['Crash Date/Time'])
    i['Month'] = i['Crash Date/Time'].dt.month
    i['Date'] = i['Crash Date/Time'].dt.day
    i['Day of Week'] = i['Crash Date/Time'].dt.dayofweek
    i['Hour'] = i['Crash Date/Time'].dt.hour

train_df = train_df.drop(['Crash Date/Time'], axis=1)
test_df = test_df.drop(['Crash Date/Time'], axis=1)
```

```python
train_df.head()
```

```python
# Splitting the 'Location' Column to 'Location1' and 'Location2' and clearing the error in the 'Location2' data

train_df[['Location1','Location2']] = train_df['Location'].str.split(pat=' |,-', expand=True)
test_df[['Location1','Location2']] = test_df['Location'].str.split(pat=' |,-', expand=True)

for row in range(len(train_df['Location2'])):
    if(train_df['Location2'][row][1]!='.'):
        train_df['Location2'][row] = train_df['Location2'][row][:1]+ "." + train_df['Location2'][row][1:]

for row in range(len(test_df['Location2'])):
    if(test_df['Location2'][row][1]!='.'):
        test_df['Location2'][row] = test_df['Location2'][row][:1]+ "." + test_df['Location2'][row][1:]

train_df = train_df.drop(['Location'], axis=1)
test_df = test_df.drop(['Location'], axis=1)
```

```python
train_df.head()
```

```python
# printing unique values under each column in train and test dataset

for col in test_df.columns:
    print(col,':',train_df[col].value_counts().count(),',',test_df[col].value_counts().count())
```

```python
train_df = train_df.drop(['Report Number','Local Case Number','Person ID','Vehicle ID'],axis=1)
test_df = test_df.drop(['Report Number','Local Case Number','Person ID','Vehicle ID'],axis=1)
```

```python
# printing number of NaN values in train and test dataset, column-wise

for col in test_df.columns:
    if(train_df[col].isnull().sum()!=0 | test_df[col].isnull().sum()!=0):
        print(col,':',train_df[col].isnull().sum(),',', test_df[col].isnull().sum())
```

```python
train_df = train_df.drop(['Cross-Street Name','Off-Road Description','Circumstance','Related Non-Motorist'], axis=1)
test_df = test_df.drop(['Cross-Street Name','Off-Road Description','Circumstance','Related Non-Motorist'], axis=1)
```

```python
train_df.head()
```

```python
values = {'Route Type': 'Unknown',
          'Road Name': 'UNKNOWN',
          'Cross-Street Type': 'Unknown',
          'Municipality': 'UNKNOWN',
          'Collision Type': train_df['Collision Type'].value_counts().index[0],
          'Weather': train_df['Weather'].value_counts().index[0],
          'Surface Condition': train_df['Surface Condition'].value_counts().index[0],
          'Light': train_df['Light'].value_counts().index[0],
          'Traffic Control': train_df['Traffic Control'].value_counts().index[0],
          'Driver Substance Abuse': 'UNKNOWN',
          'Non-Motorist Substance Abuse': 'UNKNOWN',
          'Drivers License State': train_df['Drivers License State'].value_counts().index[0],
          'Vehicle Damage Extent': 'UNKNOWN',
          'Vehicle First Impact Location': 'UNKNOWN',
          'Vehicle Second Impact Location': 'UNKNOWN',
          'Vehicle Body Type': 'UNKNOWN',
          'Vehicle Movement': 'UNKNOWN',
          'Vehicle Continuing Dir': 'Unknown',
          'Vehicle Going Dir': 'Unknown',
          'Vehicle Make': 'UNKNOWN',
          'Vehicle Model': 'UNKNOWN',
          'Equipment Problems': 'UNKNOWN',}

train_df = train_df.fillna(value = values)
test_df = test_df.fillna(value = values)

# Now the data doesn't have any NaN values
```

```python
for i in range(len(train_df)):
    if(train_df['Vehicle Year'][i]==0):
        train_df['Vehicle Year'][i]=train_df['Vehicle Year'].mode()
```

```python
train_df.head()
```

```python
train_df['Location1'] = train_df['Location1'].astype(float)
train_df['Location2'] = train_df['Location2'].astype(float)
test_df['Location1'] = test_df['Location1'].astype(float)
test_df['Location2'] = test_df['Location2'].astype(float)
```

```python
train_df.dtypes
```

```python
print(train_df.columns)
```

```python
categorical_col = ['Agency Name', 'ACRS Report Type', 'Route Type', 'Road Name',
       'Cross-Street Type', 'Municipality', 'Collision Type', 'Weather',
       'Surface Condition', 'Light', 'Traffic Control',
       'Driver Substance Abuse', 'Non-Motorist Substance Abuse',
       'Injury Severity', 'Drivers License State', 'Vehicle Damage Extent',
       'Vehicle First Impact Location', 'Vehicle Second Impact Location',
       'Vehicle Body Type', 'Vehicle Movement', 'Vehicle Continuing Dir',
       'Vehicle Going Dir', 'Driverless Vehicle',
       'Parked Vehicle', 'Vehicle Make', 'Vehicle Model',
       'Equipment Problems']

numerical_col = ['Speed Limit', 'Vehicle Year', 'Latitude', 'Longitude', 'Month', 'Date',
       'Day of Week', 'Hour', 'Location1', 'Location2']
```

```python
# removing the outliers

from scipy.stats import iqr

for col in numerical_col:
    x = iqr(train_df[col])
    Q1 = np.quantile(train_df[col],0.25)
    Q3 = np.quantile(train_df[col],0.75)
    min_ol = Q1 - 1.5*x
    max_ol = Q3 + 1.5*x
    train_df = train_df[(train_df[col]>=min_ol) & (train_df[col]<=max_ol)]

len(train_df)
```

```python
train_fault = train_df['Fault']
train_df = train_df.drop(['Fault'], axis=1)
```

```python
train_df['train']=1
test_df['train']=0

combined = pd.concat([train_df, test_df])
```

```python
for col in categorical_col:
    combined_encoded = pd.get_dummies(combined[col], sparse=True)
    combined_encoded.columns = col + '_' + combined_encoded.columns
    combined = combined.drop([col],axis=1)
    combined = pd.concat([combined, combined_encoded ], axis=1)
```

```python
combined.head()
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

for col in numerical_col:
    combined[col] = scaler.fit_transform(combined[[col]])
```

```python
combined.head()
```

```python
train_df = combined[combined['train']==1]
test_df = combined[combined['train']==0]
train_df.drop(['train'], axis=1, inplace=True)
test_df.drop(['train'], axis=1, inplace=True)
```

```python
len(train_df), len(test_df)
```

```python
train_df['Fault'] = train_fault
```

```python
# splitting training data for validtion

from sklearn.model_selection import train_test_split

training_df, valid_df = train_test_split(train_df, test_size = 0.2)

# splitting data and labels

training_labels_df = training_df['Fault']
valid_labels_df = valid_df['Fault']

training_df = training_df.drop(['Fault'], axis=1)
valid_df = valid_df.drop(['Fault'], axis=1)
```

Now, the data is pre-processed. We can proceed for modelling.

```python
from sklearn.metrics import accuracy_score
```

```python
# 1) Logistic Regression

from sklearn.linear_model import LogisticRegression

log_reg = LogisticRegression(random_state=0, max_iter=1000, solver='saga').fit(training_df, training_labels_df)
valid_predictions = log_reg.predict(valid_df)
valid_accuracy = accuracy_score(valid_labels_df, valid_predictions)
print('Logistic Regression Validation Accuracy :',valid_accuracy)

training_predictions = log_reg.predict(training_df)
training_accuracy = accuracy_score(training_labels_df, training_predictions)
print('Training Accuracy :',training_accuracy)
```

```python
# 2) Random Forest Classifier

from sklearn.ensemble import RandomForestClassifier

ran_for = RandomForestClassifier(n_estimators=800, random_state=0).fit(training_df, training_labels_df)
valid_predictions = ran_for.predict(valid_df)
valid_accuracy = accuracy_score(valid_labels_df, valid_predictions)
print('Random Forest Validation Accuracy :',valid_accuracy)

training_predictions = ran_for.predict(training_df)
training_accuracy = accuracy_score(training_labels_df, training_predictions)
print('Training Accuracy :',training_accuracy)
```

```python
# 3) Gradient Boosting Classifier

from sklearn.ensemble import GradientBoostingClassifier

grad_boost = GradientBoostingClassifier(n_estimators=200, learning_rate=0.5, max_depth=3, random_state=0).fit(training_df, training_labels_df)
valid_predictions = grad_boost.predict(valid_df)
valid_accuracy = accuracy_score(valid_labels_df, valid_predictions)
print('Gradient Boosting Classifier Validation Accuracy :',valid_accuracy)

training_predictions = grad_boost.predict(training_df)
training_accuracy = accuracy_score(training_labels_df, training_predictions)
print('Training Accuracy :',training_accuracy)
```

So, Gradient Boosting is the best model for classification here.

```python
id = np.zeros(len(test_df))
for i in range(len(test_df)):
    id[i] = i
test_result = pd.DataFrame(id, columns = ['Id'])
test_result['Id'] = test_result['Id'].astype(int)
```

```python
# prediction using Gradient Boosting Classifier
test_predictions = grad_boost.predict(test_df)
test_result['Fault'] = test_predictions
test_result.to_csv('test_result_gradboost.csv', index=False)
```
